```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data
data = pd.read_csv('../Data/customer_segment_data.csv')
data.head(5)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# what is the age distribution of customers
# How are customers distributed by gender and region
plt.figure(figsize=(8,5))
sns.histplot(data['Age'],bins=20,kde=True,color='skyblue')
plt.title('Age Distributon of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Gender'].value_counts()
```

```python
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
region_counts = data['Region'].value_counts()
plt.figure(figsize=(6, 4))
region_counts.plot(kind='bar', color='green')
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# 1. What is the average monthly revenue across regions?
2. Which region generates the highest CLTV?
```

```python
region_avg_revenue = data.groupby('Region')['Monthly_Revenue'].mean()

plt.figure(figsize=(8, 5))
region_avg_revenue.sort_values().plot(kind='bar', color='orange')
plt.title('Average Monthly Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Average Revenue ($)')
plt.show()
```

```python
region_avg_cltv  = data.groupby('Region')['CLTV'].mean()
plt.figure(figsize=(8, 5))
region_avg_cltv.sort_values().plot(kind='bar', color='purple')
plt.title('Average CLTV by Region')
plt.xlabel('Region')
plt.ylabel('Average CLTV ($)')
plt.show()
```

```python
# 1. How does data usage vary by subscription type?
# 2. What is the distribution of feedback scores?
```

```python
data['Subscription_Type'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='Subscription_Type', y='Avg_Monthly_Data_Usage_GB')
plt.title('Data Usage by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Average Data Usage (GB)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. What is the overall churn rate?
# 2. How do churned customers differ from non-churned customers?
```

```python
churn_data = data['Churn_Status'].value_counts(normalize=True)*100
plt.figure(figsize=(6, 4))
churn_data.plot(kind='pie', autopct='%1.1f%%',
                  startangle=90, colors=['orange', 'lightblue'])

plt.title('Churn Status Distribution')
plt.ylabel('')
plt.show()
```

```python
churned = data[data['Churn_Status'] == 'Yes']
non_churned = data[data['Churn_Status'] == 'No']


plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='Churn_Status', y='Monthly_Revenue', palette='Set2')
plt.title('Monthly Revenue: Churned vs Non-Churned Customers')
plt.xlabel('Churn Status')
plt.ylabel('Monthly Revenue ($)')
plt.show()
```

# Recommendations
